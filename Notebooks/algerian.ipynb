```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
ds=pd.read_csv('Algerian_forest_fires_dataset_UPDATE.csv',header=1)
```

```python
ds.head()
```

```python
ds.info()
```

Data cleaning 

```python
## missing values
ds[ds.isnull().any(axis=1)]
```

making two coloumns for two separate regions


```python
ds.loc[:122, 'region'] = 0
ds.loc[122:, 'region'] = 1
df=ds
```

```python
df[['region']]=df[['region']].astype('int')
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df=df.dropna().reset_index(drop=True)
```

```python
df.isnull().sum()
```

```python
df.iloc[[122]]
```

```python
df=df.drop(122).reset_index(drop=True)
```

```python
df.iloc[[122]]
```

```python
df.columns
```

```python
## fic spaces in columns names
```

```python
df.columns=df.columns.str.strip()
df.columns
```

change columns as intergers


```python
df[['month', 'day', 'year', 'Temperature','RH','Ws']]=df[['month', 'day', 'year', 'Temperature','RH','Ws']].astype('int')
```

```python
df.info()
```

```python
objects=[features for features in df.columns if df[features].dtype=='O']
```

```python
for i in objects:
    if i!='Classes':
        df[i]=df[i].astype(float)
```

```python
df.info()
```

```python
df.describe()
```

```python
df.to_csv('Algerian_forest_fires_cleaned.csv',index=False)
```

```python
## drop day,month and year
df_copy=df.drop(['day','month','year'],axis=1)
```

```python
df['Classes']=np.where(df['Classes'].str.contains('not fire'),'not fire','fire')
```

```python
## Monthly Fire Analysis
dftemp=df.loc[df['region']==1]
plt.subplots(figsize=(13,6))
sns.set_style('whitegrid')
sns.countplot(x='month',hue='Classes',data=df)
plt.ylabel('Number of Fires',weight='bold')
plt.xlabel('Months',weight='bold')
plt.title("Fire Analysis of Sidi- Bel Regions",weight='bold')
```
